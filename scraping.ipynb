```python
import requests
from bs4 import BeautifulSoup as bs
import urllib.request
import os
```

## Scraping System for UMASS.edu

```python
def get_link(link,img):
    return link + '/' + img

def get_image(images):
        for img in images:
            try:
                url = img['src']
                url = url.strip('../')
                img_name = os.path.basename(url)
                link_ = get_link(link,url)
                print(img_name,link_,sep='\t')
                urllib.request.urlretrieve(link_, os.path.join('downloaded_images_umass', img_name))
            except:
                continue
```

```python
url = "https://vis-www.cs.umass.edu/lfw/alpha_all_1.html"
link = '/'.join(url.split('/')[:-1])

page = requests.get(url)
soup = bs(page.text, 'html.parser')
images = soup.find_all('td')
os.makedirs('downloaded_images_umass', exist_ok=True)

for img in images:
    next_a_tag = img.find_next('a')
    if next_a_tag and next_a_tag.has_attr('href'):
        img_name = next_a_tag['href']
        link_ = get_link(link,img_name)
        response = requests.get(link_)
        soup = bs(response.text,'html.parser')
        img = soup.find_all('img')
        get_image(img)
       
```

## Scraping from Wikimedia

```python
url = "https://commons.wikimedia.org/w/index.php?search=headshot&title=Special:MediaSearch&go=Go&type=image"
page = requests.get(url)
soup = bs(page.content, 'html.parser')
images = soup.find_all('img')
os.makedirs('downloaded_images_wiki', exist_ok=True)

for img in images:
    try:
        url = img['data-src']
        img_name = os.path.basename(url)
        print(img_name,url,sep='\t')
        urllib.request.urlretrieve(url, os.path.join('downloaded_images_wiki', img_name))
    except:
        continue
```
